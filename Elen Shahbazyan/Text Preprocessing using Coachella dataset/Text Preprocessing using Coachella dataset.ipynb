```python
import pandas as pd

df = pd.read_csv(
    "../../../Natural-Language-Processing/Text Preprocessing using Coachella dataset/Coachella-2015-2-DFE.csv", encoding="ISO-8859-1")
print(df.head())
```

```python
import re
import nltk
```

```python
df["hashtags"] = df["text"].apply(lambda x: re.findall(r"#\w+", str(x)))

hashtags_df = df["hashtags"].apply(lambda x: pd.Series(x))
df = pd.concat([df, hashtags_df], axis=1)

def clean_text(text):
    if not isinstance(text, str):
        return text
    #hashtag
    text = re.sub(r"#\w+", "", text)
    #username
    text = re.sub(r"@\w+", "", text)
    #emails
    text = re.sub(r"\S+@\S+\.\S+", "", text)
    #links
    text = re.sub(r"http\S+|www\S+|https\S+", "", text, flags=re.MULTILINE)

    return text

df["text"] = df["text"].apply(clean_text)

df.to_csv("cleaned_file.csv", index=False)
```

```python
def remove_non_ascii_symbols(file_path: str):
    df = pd.read_csv(file_path, dtype=str)
    df = df.applymap(lambda x: re.sub(r'[^\x00-\x7F]', '', x) if pd.notnull(x) else x)
    df.to_csv(file_path, index=False)

remove_non_ascii_symbols("cleaned_file.csv")
df
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
tweet_col = 'text'
```

```python
def to_lower(text):
    return text.lower()
```

```python
stop_words = set(stopwords.words('english'))

def remove_stop_words(text):
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return " ".join(filtered_words)
```

```python
def remove_digits(text):
    return re.sub(r'\d+', '', text)
```

```python
def remove_special_characters(text):
    return re.sub(r'[^\w\s]', '', text)
```

```python
df['cleaned_text'] = df[tweet_col].apply(to_lower)
df['cleaned_text'] = df['cleaned_text'].apply(remove_digits)
df['cleaned_text'] = df['cleaned_text'].apply(remove_special_characters)
df['cleaned_text'] = df['cleaned_text'].apply(remove_stop_words)

df.to_csv("final_cleaned_file.csv", index=False)

```
